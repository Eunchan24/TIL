```python
import nltk
from konlpy.tag import Kkma
from konlpy.tag import Okt
from collections import Counter
import matplotlib.pyplot as plt
```

```python
# 명사 분석 (nouns)

f = open('문재인.txt','r',encoding='utf-8')
news = f.read()

# okt 객체 생성
okt = Okt()

# 명사 추출
noun = okt.nouns(news)
for i,v in enumerate(noun):
    # 1글자 삭제
    if len(v) < 2 :
        noun.pop(i)
            
# 명사 빈도 카운트 및 저장(dict)        
count = Counter(noun)
# print(count)

# 명사 빈도 top100
noun_list = count.most_common(100)
print(noun_list)
# for v in noun_list:
#     print(v)
```

```python
from wordcloud import WordCloud

font_path = 'C:\\Users\\user\\han.ttf'

wc = WordCloud(font_path= font_path, 
    background_color="white", width=1000, height=1000, max_words=100, max_font_size=300)

wc.generate_from_frequencies(dict(noun_list))
# wc.to_file('wordcloud_news.png')  # 파일 저장

plt.figure(figsize=(10, 10))
plt.imshow(wc)
plt.axis("off")
plt.show()
```

```python
from PIL import Image
import numpy as np

# 이미지 로드 후 np.array 변경
korea = np.array(Image.open("presentation.png"))

# worldcloud 생성
wordcloud = WordCloud(
    font_path = font_path,
    width = 800,
    height = 800,
    background_color="white",
    mask = korea  # 마스크 이미지 적용
)

wordcloud = wordcloud.generate_from_frequencies(dict(noun_list))
plt.figure(figsize=(15, 15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
