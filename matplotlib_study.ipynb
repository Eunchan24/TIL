```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

labels = list('ABCDE')
x = np.arange(len(labels)) # 5개
y1 = np.array([1, 2, 10, 4, 3])
y2 = np.array([1, 10, 2, 5, 1])

# 하늘색 
axes[0].bar(x, y1, color='#1ABDE9')

# 분홍색
axes[1].bar(x, y1, color='#FECACA', edgecolor='black', linewidth=2)

# 확인
for idx, ax in enumerate(axes, 1):
    
    # 제목
    ax.set_title(f"Bar Plot {idx}", fontsize=15, fontweight='bold')

    # Ticklabels
    ax.set_xticks(x)
    ax.set_xticklabels(labels, fontsize=12)
    ax.tick_params(axis='y', which='major', labelsize=13)
    
    # 핑크색일때(axes[1])
    if idx==1 : continue
    ax.spines['top'].set_visible(False)  # 상단 선 안보이게 하기
    ax.spines['right'].set_visible(False)  # 오른쪽 선 안보이게 하기
    ax.spines['bottom'].set_linewidth(2)  # 하단 선 두께 2
    ax.spines['left'].set_linewidth(2)  # 왼쪽 선 두께 2
    
plt.show()
```

```python
fig = plt.figure(figsize=(16, 7), dpi=150)
np.random.seed(970725)

# 랜덤값 20개 생성
x = np.random.rand(20)
y = np.random.rand(20)

# 산점도 그리기
axes = []
for i in range(1, 3):
    ax = fig.add_subplot(1, 2, i, aspect=1)
    axes.append(ax)
    ax.scatter(x, y, s=150, c='#1ABDE9', linewidth=1.5, edgecolor='black', zorder=10)

    # 그래프 라인 설정
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_linewidth(1.5)
    ax.spines['left'].set_linewidth(1.5)

    # x, y 축 값 설정
    ax.set_xlim(0, 1.5)
    ax.set_ylim(0, 1.5)

    
axes[0].grid()  # 눈금선 표시
axes[0].set_title(f"Scatter Plot w Grid", fontsize=15, va= 'center')
axes[1].set_title(f"Scatter Plot w/o Grid", fontsize=15, va= 'center')

plt.show()
```

```python
fig = plt.figure(figsize=(7, 7), dpi=150)
ax = fig.add_subplot(111, aspect=1)

ax.scatter([1, 2, 3], [3, 3, 3], 
           color=['#1ABDE9', '#F36E8E', '#00BE6C'],
           s=300,
           linewidth=1.5,
           edgecolor='black'
          )

for idx, marker in enumerate(['o', '^', 's'], 1):
    ax.scatter(idx, 2,
           color='yellow',
           marker=marker,
           s=300,
           linewidth=1.5,
           edgecolor='blue'
          )

    
for idx, s in enumerate([300, 1200, 2700], 1):
    ax.scatter(idx, 1,
           color='purple',
           s=s,
           linewidth=1.5,
           edgecolor='pink'
          )    

# 라인 감추기    
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

ax.set_xlim(0, 4)
ax.set_ylim(0, 3.5)
ax.set_xticks([])
ax.set_yticks([])
ax.set_title(f"Scatter Components", fontsize=15, va= 'center')

plt.show()
```
